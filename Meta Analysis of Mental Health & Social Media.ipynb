```python
import pandas as pd
import numpy as np
import nltk
```

```python
dataset = pd.read_csv("/Users/deepthitamma/Desktop/file6.csv", sep="\t",encoding="ISO-8859-1",header=None)
dataset.head()
dataset.columns=['Text']
dataset
```

```python
import re
import string
```

```python
def remove_punct(text):
    text_nopunct = "".join([char for char in text if char not in string.punctuation])
    return text_nopunct

dataset['Text_clean'] = dataset['Text'].apply(lambda x: remove_punct(x))
dataset.head()
```

```python
def tokenize(text):
    tokens = re.split('\W+', text)
    return tokens

dataset['Text_tokenized'] = dataset['Text_clean'].apply(lambda x: tokenize(x.lower()))
dataset.head()
```

```python
stopword = nltk.corpus.stopwords.words('english')
stopword
def remove_stopwords(tokenized_list):
    text = [word for word in tokenized_list if word not in stopword]
    return text

dataset['Text_nostop'] = dataset['Text_tokenized'].apply(lambda x: remove_stopwords(x))

dataset.head()
```

```python
dataset.to_csv (r'/Users/deepthitamma/Desktop/A6_converted.csv', index = False, header=True)
```
